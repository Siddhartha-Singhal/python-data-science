merge - used to merge 2 or more datframes

```python
df1 = pd.DataFrame({'Roll no.': [1, 2, 3, 4, 5], 'Physics': [13, 22, 45, 33, 32]})
df2 = pd.DataFrame({'Roll no.': [1, 2, 3, 4, 5], 'Chemistry':[34, 66, 24, 63, 75]})
pd.merge(df1, df2)  # by default intersecting column is taken if 'on' is not provided
#  pd.merge(df1, df2, on='Roll no.')
```

if both tables have different values in column used for combining

```python
df1 = pd.DataFrame({'Roll no.': [1, 2, 3, 4, 5], 'Physics': [13, 22, 45, 33, 32]})
df2 = pd.DataFrame({'Roll no.': [1, 2, 3, 5, 6], 'Chemistry':[34, 66, 24, 63, 75]})
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
df1 = pd.DataFrame({'Roll no.': [1, 2, 3, 4, 5], 'Physics': [13, 22, 45, 33, 32]})
df2 = pd.DataFrame({'Roll no.': [6, 7, 8, 9, 10], 'Physics':[34, 66, 24, 63, 75]})
print(df1)
print(df2)
```

```python
pd.merge(df1, df2, how='outer').set_index('Roll no.')
```

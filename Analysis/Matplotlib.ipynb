## Scatter plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

To create a dark background

```python
plt.style.use('dark_background')
```

```python
rollNo = [1, 2, 3, 4, 5, 6, 7, 8, 9]
marks = [14, 3, 6, 17, 16, 11, 18, 20, 9]
```

scatter(x-axis, y-axis, color, marker)
- x-axis - values on x-axis
- y-axis - values on y-axis
- color - color of the markers

```python
plt.scatter(rollNo, marks, color = 'green')
plt.show()
```

scatter( , , ,marker)
- marker - shows the type of symbols to be used for scatter plot

```python
plt.scatter(rollNo, marks, color = 'red', marker='v')
plt.show()
```

figure(figsize) 
- helps to define the size of the graph
- used to plot multiple plots in same graph

```python
plt.figure(figsize=(8, 4))
plt.scatter(rollNo, marks, color = 'red', marker='v')
plt.show()
```

other function using plot

- markersize - define the size of the symbol
- 'g*' - means green color * are used as symbols

```python
plt.figure(figsize=(8,4))
plt.plot(rollNo, marks, 'g*', markersize = 10)
plt.show()
```

Multiple plots on same figure

```python
temp_pune = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_pune = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]

temp_bangalore = [34, 35, 36, 37, 28, 27, 26, 25, 31, 20]
humidity_bangalore = [40, 38, 36, 35, 42, 44, 41, 40, 34, 45]
```

- xtick - set the values on x-axis
- ytick - set the values on y-axis

```python
plt.figure(figsize=(12, 6))
plt.xticks(np.arange(0, 60, 5))
plt.yticks(np.arange(10, 60, 5))

plt.plot(temp_pune, humidity_pune, 'bo', markersize=10)
plt.plot(temp_bangalore, humidity_bangalore, 'go', markersize=10)
plt.show()
```

### Making visualisations on Iris dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
plt.scatter(df['sepal_length'], df['sepal_width'], color='b', marker='v', s=8)
plt.show()
```

- alpha - it is used to set transparency, 0 for completely transparent and 1 for opaque

```python
df.plot(kind='scatter', x='sepal_length', y='petal_length',color='green', marker='v', alpha=0.5, s=12)

plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.show()
```

## Line plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
rollNo = [1, 2, 3, 4, 5, 6, 7, 8, 9]
marks = [14, 3, 6, 17, 16, 11, 18, 20, 9]
```

```python
plt.plot(rollNo, marks, linestyle='--', color='purple')
plt.show()
```

```python
plt.plot(rollNo, marks, linestyle='-', color='red', linewidth=5)
plt.show()
```

```python
plt.plot(rollNo, marks, linestyle=':', color='pink', linewidth=2)
plt.show()
```

```python
plt.plot(rollNo, marks, linestyle='-.', color='orange', linewidth=3)
plt.show()
```

```python
plt.plot(rollNo, marks, linestyle='-.', color='yellow')
plt.show()
```

Multiple plots on same graph

```python
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
```

```python
plt.figure(figsize=(8,8))
plt.xticks(np.arange(0,15,1))
plt.yticks(np.arange(0,101,5))

plt.plot(study_hours, marks, linestyle = '-', color = 'green', linewidth = 2)
plt.plot(study_hours, marks, 'bo', markersize = 6)

plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```
